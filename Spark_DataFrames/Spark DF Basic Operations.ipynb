## Spark DataFrame Basic Operations

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
```

```python
spark = SparkSession \
        .builder \
        .appName('ops') \
        .getOrCreate()
```

```python
df = spark.read.csv('appl_stock.csv', inferSchema = True, header = True)
```

```python
df.printSchema()
```

```python
df.show()
```

### Filtering the data

```python
df.filter("Close > 200").show()
```

```python
df.filter("Close < 500").select(['Open', 'Close']).show()
```

```python
# Filtering on multiple booleans
df.filter((df['Close'] < 200) & (df['Open'] > 200)).show()
```

```python
result = df.filter(df['Low'] == 197.16).collect() # Find the date
result[0][0]
```

```python
row = result[0]
row.asDict()['Date']
```
